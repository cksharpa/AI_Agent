```python
from dotenv import load_dotenv
```

```python
import os

ollama_api_key = os.getenv('OLLAMA_API_KEY')

if ollama_api_key:
    print(f"Ollama API Key exists and begins {ollama_api_key[:8]}")
else:
    print("Ollama API Key not set - please head to the troubleshooting guide in the setup folder")

print(os.getenv('OLLAMA_HOST'))
print(os.getenv('OLLAMA_MODEL'))
```

```python
 messages=[{"role": "user", "content": "Hello, how are you?"}]
```

```python
import ollama

response = ollama.chat(
    model="llama3.2",
    messages=messages
)

print(response.message.content)
```

```python
question = "Suggest a hard and challenging question to assess someone's IQ. Respond only with the question."
messages = [{"role": "user", "content": question}]
```

```python
response = ollama.chat(
    model="llama3.2",
    messages=messages
)

question = response.message.content

print(question)
```
