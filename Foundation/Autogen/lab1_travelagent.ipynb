```python
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_ext.models.ollama import OllamaChatCompletionClient
from autogen_agentchat.messages import TextMessage
from autogen_agentchat.agents import AssistantAgent
from autogen_core import CancellationToken
from dotenv import load_dotenv
import os
```

```python
load_dotenv(override=True)
```

```python
# Need to provide model_info if using opensource models from openrouter like gpt-oss-20b
# else we can just use OpenAIChatCompletionClient(model="gpt-4o-mini")
gpt_client = OpenAIChatCompletionClient(
    model="openrouter/gpt-oss-20b",
    api_key=os.environ["OPENAI_API_KEY"],
    base_url=os.environ["OPENAI_BASE_URL"],
    model_info={
        "family": "openai",
        "vision": False,
        "function_calling": True,
        "json_output": True,
    },
)
```

```python
#Ollama model usage
ollama_client = OllamaChatCompletionClient(model="llama3.2")
```

```python
message = TextMessage(content="Plan 14 days trip to North-east India", source="user")
message
```

```python
#Agent
agent = AssistantAgent(name="tour_manager", model_client=ollama_client,
    system_message="You are helpful travel assistant. You plan day-by-day itinerary and reply user in witty but respectful way.",
    model_client_stream=True
)
```

```python
response = await agent.on_messages([message], cancellation_token=CancellationToken())
response.chat_message.content
```
