# Communication using Standalone Runtime

```python
from dataclasses import dataclass
from autogen_core import AgentId, MessageContext, RoutedAgent, message_handler
from autogen_core import SingleThreadedAgentRuntime
from autogen_agentchat.agents import AssistantAgent
from autogen_agentchat.messages import TextMessage
from autogen_ext.models.openai import OpenAIChatCompletionClient
from autogen_ext.models.ollama import OllamaChatCompletionClient
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
```

```python
#Defining message object

@dataclass
class Message:
    content: str
```

```python
#Defining Agent
class TestAgent(RoutedAgent):
    def __init__(self) -> None:
        super().__init__("Test")

    @message_handler
    async def on_message(self, message: Message, ctx:MessageContext) -> Message:
        return Message(content=f"This is {self.id.type}-{self.id.key}. You responded '{message.content} and It's fine.")
```

```python
#Creating Standalone Runtime
runtime = SingleThreadedAgentRuntime()
await TestAgent.register(runtime, "test_agent", lambda: TestAgent())
```

```python
#Starting runtime 
runtime.start()
```

```python
# Sending message
id = AgentId("test_agent", "default")
response = await runtime.send_message(Message("How are you ?"), id)
print("--->", response.content)
```

```python
await runtime.stop()
await runtime.close()
```
