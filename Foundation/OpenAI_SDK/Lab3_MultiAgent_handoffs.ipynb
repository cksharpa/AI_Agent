```python
from dotenv import load_dotenv
from agents import Agent, Runner, function_tool
import asyncio
```

```python
load_dotenv(override=True)
```

```python
math_expert_agent = Agent(
    
name="Math Expert",
    
handoff_description="Expert agent for math questions",
    
instructions="You answer the math questions in concise manner", model="gpt-oss-20b")
```

```python
english_expert_agent = Agent(
    
name="English Expert",
    
handoff_description="Specialist agent for English questions",
    
instructions="You answer English and grammer related question in concise manner.", model="gpt-oss-20b")
```

```python
homework_agent = Agent(
    name=" Homework Agent",
    instructions="Route each homework question to the right Expert.",
    handoffs=[math_expert_agent, english_expert_agent],
)
```

```python
result = await Runner.run(
homework_agent, "Active to Passive: I will clean the house.",)
    
print(result.final_output)
    
print(f"Answered by: {result.last_agent.name}")
```
