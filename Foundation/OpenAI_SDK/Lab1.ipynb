```python
from dotenv import load_dotenv
from agents import Agent, Runner
import asyncio
import os
```

```python
load_dotenv(override=True)
```

```python
#Setting up the API keys. Set OPENAI_API_KEY same as OPENROUTER_API_KEY
#and OPENAI_BASE_URL="https://openrouter.ai/api/v1" in environment variables(.env file)
ollama_api_key = os.getenv('OLLAMA_API_KEY')
openrouter_api_key = os.getenv('OPENROUTER_API_KEY')
openai_api_key = os.getenv('OPENAI_API_KEY')

if ollama_api_key:
    print(f"Ollama API Key exists and begins {ollama_api_key[:8]}")
else:
    print("Ollama API Key not set")
    
if openrouter_api_key:
    print(f"Openrouter API Key exists and begins {openrouter_api_key[:7]}")
else:
    print("Openrouter API Key not set")

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
```

```python
agent = Agent(name="Math Tutor", 
instructions="You answer math questions clearly and concisely.", model="gpt-oss-20b"
)
```

```python
result = await Runner.run(agent, "What is the square root of 14884?")
print(result.final_output)
```
