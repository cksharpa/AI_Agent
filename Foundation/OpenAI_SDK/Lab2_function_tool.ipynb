```python
from dotenv import load_dotenv
from agents import Agent, Runner, function_tool
import asyncio
```

```python
load_dotenv(override=True)
```

```python
@function_tool
def interesting_fact():
    return "Interesting fact : From 0 to 1000, the only number that has the letter “a” in it is “one thousand”"
```

```python
agent = Agent(name="Math Teacher", 
instructions="You are a helpful math teacher that can answer math questions clearly, Use interesting_fact to tell interesting facts about numbers.",
model="gpt-oss-20b",
tools=[interesting_fact], )
```

```python
result = await Runner.run(agent, "Tell me the interesting fact about numbers")
print(result.final_output)
```
